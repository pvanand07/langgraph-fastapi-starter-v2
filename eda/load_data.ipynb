```python
# Parse the Excel file to extract metadata and load the table
import pandas as pd
import numpy as np
from IPython.display import display

excel_file_path = "business_data/All Product Purchase Report.xlsx"

def perform_eda(excel_file_path):


    # Read the raw Excel file without header to see all rows
    raw_df = pd.read_excel(excel_file_path, header=None)

    # Debug: Check the first few rows and columns
    print("Debug - First 6 rows, first column:")
    print(raw_df.iloc[0:6, 0])
    print("\nDebug - Column names:", raw_df.columns.tolist()[:5])
    print()

    # Extract metadata from rows 0-5, first column (index 0)
    metadata = {}
    if len(raw_df) > 0:
        # Get the first column (should be index 0 when header=None)
        first_col = raw_df.iloc[:, 0]
        
        # Row 0: Company name (e.g., 'MARWA ENTERPRISES')
        val_0 = first_col.iloc[0]
        if pd.notna(val_0):
            metadata['name'] = str(val_0).strip()
        
        # Row 1: Address (e.g., 'ALUVA: 683101')
        val_1 = first_col.iloc[1]
        if pd.notna(val_1):
            metadata['address'] = str(val_1).strip()
        
        # Row 2: Empty (NaN)
        # Skip row 2
        
        # Row 3: Report title (e.g., 'All Product Purchase Report')
        val_3 = first_col.iloc[3]
        if pd.notna(val_3):
            metadata['report_title'] = str(val_3).strip()
        
        # Row 4: Company (e.g., 'Company :MONDELEZ INDIA FOODS LTD')
        val_4 = first_col.iloc[4]
        if pd.notna(val_4):
            company_str = str(val_4).strip()
            if 'Company' in company_str:
                metadata['client'] = company_str.replace('Company', '').replace(':', '').strip()
            else:
                metadata['client'] = company_str
        
        # Row 5: Date range (e.g., 'From: 01/04/2025  To: 28/11/2025')
        val_5 = first_col.iloc[5]
        if pd.notna(val_5):
            date_str = str(val_5).strip()
            metadata['date_range'] = date_str
            # Extract individual dates
            if 'From:' in date_str and 'To:' in date_str:
                try:
                    from_date = date_str.split('From:')[1].split('To:')[0].strip()
                    to_date = date_str.split('To:')[1].strip()
                    metadata['from_date'] = from_date
                    metadata['to_date'] = to_date
                except Exception as e:
                    pass

    # Load the table starting from row 6 (which contains column headers)
    # Skip rows 0-5 and use row 6 as header
    df = pd.read_excel(excel_file_path, header=6)

    # Use the first row as column names (it contains the actual column headers)
    if len(df) > 0:
        # Get column names from first row
        df.columns = df.iloc[0].values
        # Drop the first row (it was the header row)
        #df = df.drop(df.index[0])

    # Clean up the DataFrame - remove any rows that are completely NaN
    print("shape of df before cleaning", df.shape)

    df = df.dropna(how='all')

    # Reset index
    df = df.reset_index(drop=True)

    # Display results
    print("Metadata:")
    for key, value in metadata.items():
        print(f"  {key}: {value}")
    print("\nDataFrame shape:", df.shape)
    print("\nDataFrame columns:", df.columns.tolist())
    print("\nFirst few rows:")
    display(df.head())


perform_eda(excel_file_path)
```

```python
import pandas as pd
import numpy as np
import duckdb
import hashlib
import re
from pathlib import Path
from IPython.display import display

# Initialize DuckDB connection
conn = duckdb.connect()

def generate_table_name(file_path):
    """
    Generate a table name from filename + file hash.
    Format: filename_lowercase_letters_only_hash5
    Only keeps letters (a-z), removes numbers, special characters, and stop words.
    """
    # Common stop words to remove
    stop_words = {
        'a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from',
        'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the',
        'to', 'was', 'were', 'will', 'with', 'wise', 'all', 'any', 'both',
        'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor',
        'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 'can',
        'could', 'should', 'would', 'may', 'might', 'must', 'shall'
    }
    
    # Get filename without extension
    path_obj = Path(file_path)
    filename = path_obj.stem  # filename without extension
    
    # Generate file hash
    with open(file_path, 'rb') as f:
        file_hash = hashlib.md5(f.read()).hexdigest()[:5]
    
    # Convert to lowercase and keep ONLY letters (a-z), replace everything else with underscore
    # This removes numbers, spaces, hyphens, and all special characters
    name_part = re.sub(r'[^a-z]', '_', filename.lower())
    
    # Split by underscores and filter out stop words
    words = [word for word in name_part.split('_') if word and word not in stop_words]
    
    # Join words back with underscores
    name_part = '_'.join(words)
    
    # Remove multiple consecutive underscores and strip underscores from ends
    name_part = re.sub(r'_+', '_', name_part).strip('_')
    
    # If name_part is empty after removing all non-letters and stop words, use a default
    if not name_part:
        name_part = 'table'
    
    # Combine: filename_hash
    table_name = f"{name_part}_{file_hash}"
    
    return table_name

def load_excel_to_dataframe(excel_file_path):
    """
    Load an Excel file and extract both metadata and data table.
    Returns a tuple: (metadata_dict, cleaned_dataframe)
    """
    # Read the raw Excel file without header to see all rows
    raw_df = pd.read_excel(excel_file_path, header=None)
    
    # Extract metadata from rows 0-5, first column (index 0)
    metadata = {}
    if len(raw_df) > 0:
        first_col = raw_df.iloc[:, 0]
        
        # Row 0: Company name
        val_0 = first_col.iloc[0]
        if pd.notna(val_0):
            metadata['name'] = str(val_0).strip()
        
        # Row 1: Address
        val_1 = first_col.iloc[1]
        if pd.notna(val_1):
            metadata['address'] = str(val_1).strip()
        
        # Row 3: Report title
        val_3 = first_col.iloc[3]
        if pd.notna(val_3):
            metadata['report_title'] = str(val_3).strip()
        
        # Row 4: Company/Client
        val_4 = first_col.iloc[4]
        if pd.notna(val_4):
            company_str = str(val_4).strip()
            if 'Company' in company_str:
                metadata['client'] = company_str.replace('Company', '').replace(':', '').strip()
            else:
                metadata['client'] = company_str
        
        # Row 5: Date range
        val_5 = first_col.iloc[5]
        if pd.notna(val_5):
            date_str = str(val_5).strip()
            metadata['date_range'] = date_str
            if 'From' in date_str and 'To' in date_str:
                try:
                    if 'From:' in date_str:
                        from_date = date_str.split('From:')[1].split('To:')[0].strip()
                        to_date = date_str.split('To:')[1].strip()
                    elif 'From Date' in date_str:
                        from_date = date_str.split('From Date')[1].split('To')[0].replace(':', '').strip()
                        to_date = date_str.split('To Dt:')[1].strip()
                    else:
                        from_date = date_str.split('From')[1].split('To')[0].strip()
                        to_date = date_str.split('To')[1].strip()
                    metadata['from_date'] = from_date
                    metadata['to_date'] = to_date
                except Exception as e:
                    pass
    
    # Load the table starting from row 6 (which contains column headers)
    df = pd.read_excel(excel_file_path, header=6)
    
    # Use the first row as column names
    if len(df) > 0:
        df.columns = df.iloc[0].values
        # Drop the first row (it was the header row)
        df = df.drop(df.index[0])
    
    # Clean up the DataFrame
    df = df.dropna(how='all')
    df = df.reset_index(drop=True)
    
    # Remove any rows where the first column contains the header names (duplicate headers)
    if len(df) > 0:
        first_col_name = df.columns[0]
        # Remove rows where first column equals the column name itself
        df = df[df[first_col_name] != first_col_name]
        df = df.reset_index(drop=True)
    
    # Convert date columns to DATE type if they exist
    date_columns = ['Date', 'from_date', 'to_date']  # Common date column names
    for col in date_columns:
        if col in df.columns:
            try:
                # Try to parse dates in DD/MM/YYYY format
                df[col] = pd.to_datetime(df[col], format='%d/%m/%Y', errors='coerce')
            except:
                try:
                    # Fallback to auto-detect format
                    df[col] = pd.to_datetime(df[col], errors='coerce')
                except:
                    pass  # Keep as string if conversion fails
    
    return metadata, df

# Load each Excel file
excel_files = [
    'business_data/All Product Purchase Report.xlsx',
    'business_data/Salesman Wise PARTY Wise Summary 1-04-2025 to 28-11-25.xlsx',
    'business_data/Salesman Wise Product Summary.xlsx'
]

dataframes = {}
metadata_dict = {}
file_to_table = {}  # Map file paths to table names

for file_path in excel_files:
    print(f"\n{'='*60}")
    print(f"Loading: {file_path}")
    print(f"{'='*60}")
    
    try:
        # Generate table name automatically
        table_name = generate_table_name(file_path)
        file_to_table[file_path] = table_name
        
        print(f"Generated table name: {table_name}")
        
        metadata, df = load_excel_to_dataframe(file_path)
        metadata_dict[table_name] = metadata
        dataframes[table_name] = df
        
        print(f" Successfully loaded {table_name}")
        print(f"  Shape: {df.shape}")
        print(f"  Columns: {list(df.columns)}")
        print(f"  Metadata: {metadata}")
        
        # Create DuckDB table
        conn.register(table_name, df)
        print(f" Created DuckDB table: {table_name}")
        
    except Exception as e:
        print(f" Error loading {file_path}: {str(e)}")
        import traceback
        traceback.print_exc()

# Create metadata table
metadata_rows = []
for table_name, meta in metadata_dict.items():
    # Find the source file for this table
    source_file = None
    for file_path, tbl_name in file_to_table.items():
        if tbl_name == table_name:
            source_file = file_path
            break
    
    row = {
        'table_name': table_name,
        'source_file': source_file,
        'name': meta.get('name', None),
        'address': meta.get('address', None),
        'report_title': meta.get('report_title', None),
        'client': meta.get('client', None),
        'date_range': meta.get('date_range', None),
        'from_date': meta.get('from_date', None),
        'to_date': meta.get('to_date', None),
        'row_count': len(dataframes[table_name]),
        'column_count': len(dataframes[table_name].columns)
    }
    metadata_rows.append(row)

metadata_df = pd.DataFrame(metadata_rows)
conn.register('metadata', metadata_df)
print(f" Created DuckDB table: metadata")

print(f"\n{'='*60}")
print("Summary:")
print(f"{'='*60}")
print(f"Total tables created: {len(dataframes) + 1} (including metadata)")
print(f"\nAvailable tables in DuckDB:")
for table_name in dataframes.keys():
    result = conn.execute(f"SELECT COUNT(*) as row_count FROM {table_name}").fetchone()
    print(f"  - {table_name}: {result[0]} rows")
print(f"  - metadata: {len(metadata_df)} rows")

print(f"\nMetadata table contents:")
print("=" * 60)
display(metadata_df)
```
